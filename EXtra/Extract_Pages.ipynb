```python
from pypdf import PdfReader, PdfWriter

in_path = 'Microelectronic Circuits (The Oxford Series in Electrical -- Adel S. Sedra, Kenneth C. Smith -- ( WeLib.org ) (1).pdf'
out_path = 'first_14_pages.pdf'

N = 14

reader = PdfReader(in_path)
writer = PdfWriter()
num_pages = len(reader.pages)
end = min(N,num_pages)

for i in range(end):
    writer.add_page(reader.pages[i])

with open(out_path, 'wb') as f:
    writer.write(f)
print(f"saved {end} pages to: {out_path}")
```

```python
from pypdf import PdfReader, PdfWriter

in_path = 'Rudy J. van de Plassche - CMOS Integrated Analog-to-Digital and Digital-to-Analog Converters (The Springer International Series in Engineering and Comput (2010, Springer) - libgen.li.pdf'
out_path = 'Flash_ADC.pdf'

# 1-indexed page numbers you want
pages_to_extract = [159,160, 161]

reader = PdfReader(in_path)
writer = PdfWriter()
num_pages = len(reader.pages)

for p in pages_to_extract:
    idx = p - 1  # convert to 0-index
    if 0 <= idx < num_pages:
        writer.add_page(reader.pages[idx])
    else:
        raise ValueError(f"Requested page {p} is out of range (1..{num_pages}).")

with open(out_path, 'wb') as f:
    writer.write(f)

print(f"saved {len(pages_to_extract)} pages to: {out_path}")
```
